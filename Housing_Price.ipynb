**In order to replace these remaining missing values, it is important to check which columns contain the numerical (integer and float) values and which columns contain 'object' type values**

```python
# Obtaining the columns containing integer and floating point values

df.select_dtypes(include=['int64', 'float64']).columns
```

```python
# Obtaining the columns containing 'object' values

df.select_dtypes(include=['object']).columns
```

**Numeric** type missing values columns are 'LotFrontage','MasVnrArea','GarageYrBlt',

**Object** type missing values columns are 'GarageFinish','GarageQual','GarageCond','MasVnrType', 'BsmtQual','BsmtCond','BsmtExposure', 'BsmtFinType1','BsmtFinType2','Electrical','GarageType',

# üü•üü• Phase B: What Will Happen? (AI Models)

### Getting dummies for categorical variables

```python
# Selecting the columns for which dummy variables are required
column_for_dummies = ['MSZoning', 'Street', 'LotShape', 'LandContour', 'Utilities',
       'LotConfig', 'LandSlope', 'Neighborhood', 'Condition1', 'Condition2',
       'BldgType', 'HouseStyle', 'RoofStyle', 'RoofMatl', 'Exterior1st',
       'Exterior2nd', 'MasVnrType', 'ExterQual', 'ExterCond', 'Foundation',
       'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2',
       'Heating', 'HeatingQC', 'CentralAir', 'Electrical', 'KitchenQual',
       'Functional', 'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
       'PavedDrive', 'SaleType', 'SaleCondition']

# Getting dummies for categorical variables and storing the result in a new dataframe called final_df
final_df = pd.get_dummies(df, columns=column_for_dummies, drop_first=True)
```

```python
# Checking the shape of final dataframe after getting dummy variables

final_df.shape
```

```python
# Checking the head of final dataframe

final_df.head()
```

```python
# Checking the data types for all the columns of final dataframe to check whether all the columns are numeric or not

pd.set_option("display.max_rows", 300) #this will display all the rows in the result

final_df.dtypes
```

```python
# The 'ID' column will not provide any meaningful information of House Prices so it's better to drop it

final_df.drop(['Id'], axis=1, inplace=True)
```

```python
# Checking the shape of final dataframe

final_df.shape
```

## 5Ô∏è‚É£ Step 5: Predictive Modeling

### Splitting the data into training set and test set

```python
# Assigning the input features to 'X' and target column to 'y'

X = final_df.drop('SalePrice', axis=1).values
y = final_df['SalePrice'].values
```

```python
# importing required library
from sklearn.model_selection import train_test_split

#Splitting the data into training and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

### Training model

```python
# importing the required library
from sklearn.ensemble import RandomForestRegressor

# defining the model
model = RandomForestRegressor()

#fitting the model to training data
model.fit(X_train,y_train)

print('Accuracy of Random Forest regressor on training set: {:.2f}'.format(model.score(X_train, y_train))) #accuracy on training data
print('Accuracy of Random Forest regressor on test set: {:.2f}'.format(model.score(X_test, y_test))) #accuracy on test data
```

### Making predictions from the model

```python
# Making predictions on test set

y_pred = model.predict(X_test)
```

```python
# getting the values of actual test data and predicted values and putting them in a dataframe called 'result_df' to compare them side by side

result_df = pd.DataFrame({'Actual Data':y_test, 'Predictions':y_pred})
print(result_df)
```

```python
# Plotting the actual test values vs. the predicted values by model

sns.scatterplot(y_test,y_pred)
```

```python
# Plotting the difference between the Test data set and predicted valuesby model
# We expect the difference to be normally distributed aroud zero if the model is making good predictions

sns.distplot(y_test-y_pred)
```

# üü•üü•üü• Phase C: Are You Sure? (Model Evaluation)

## 6Ô∏è‚É£ Step 6: Evaluation

```python
# Importing required performance metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error

# Getting the values of various performance metrics
print('Mean Absolute Error: {:.2f}'.format(mean_absolute_error(y_test, y_pred)))
print('Mean Squared Error:{:.2f}'.format(mean_squared_error(y_test, y_pred)))
print('Root Mean Squared Error:{:.2f}'.format(np.sqrt(mean_squared_error(y_test, y_pred))))
```

```python
# Checking the R-squared (R2) score

# Importing the required library
from sklearn.metrics import r2_score

# Getting the R2 score
print('R2: {:.2f}'.format(r2_score(y_test, y_pred)))
```
